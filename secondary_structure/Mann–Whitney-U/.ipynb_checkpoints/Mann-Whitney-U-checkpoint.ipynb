```python
import random
from scipy.stats import mannwhitneyu
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
################################################################ Secondary structural plot for Pichia datasets
PPD=pd.read_csv('PpD_structure.txt',sep='\t')
PPE=pd.read_csv('PpE_structure.txt',sep='\t')
PPD['data']='depleted_fragments'
PPE['data']='enriched_fragments'
PPD=PPD.drop(['ID'],axis=1)
PPE=PPE.drop(['ID'],axis=1)

cdf = pd.concat([PPD,PPE]) 
mdf = pd.melt(cdf, id_vars=['data'], var_name=['Secondary structural class'])
plt.figure(figsize=(15,10))

sns.boxplot(x="Secondary structural class", y="value", hue='data',data=mdf,palette=["orange","darkcyan"]) 
plt.legend(loc='upper right')
plt.title("% of Secondary structural elements")
```

```python
################################################################# Mannwhitney-U  stats for Pichia datasets

helix=mannwhitneyu(PPD['H'], PPE['H'], alternative='two-sided')
sheet=mannwhitneyu(PPD['E'], PPE['E'], alternative='two-sided')
coil=mannwhitneyu(PPD['C'], PPE['C'], alternative='two-sided')

print ("Mann whitney stat for helices: ",helix,'\n')
print ("Mann whitney stat for strands: ",sheet,'\n')
print ("Mann whitney stat for coils: ",coil,'\n')
```

```python
############################################################ Secondary structural plot for Saccharomyces datasets

SCD=pd.read_csv('ScD_structure.txt',sep='\t')
SCE=pd.read_csv('ScE_structure.txt',sep='\t')
SCD['data']='depleted'
SCE['data']='enriched'
SCD=SCD.drop(['ID'],axis=1)
SCE=SCE.drop(['ID'],axis=1)

cdf = pd.concat([SCD,SCE]) 
mdf = pd.melt(cdf, id_vars=['data'], var_name=['Secondary structural class'])
plt.figure(figsize=(15,10))

sns.boxplot(x="Secondary structural class", y="value", hue='data',data=mdf,palette=["orange","darkcyan"]) 
plt.legend(loc='upper right')
plt.title("% of Secondary structural elements")
```
